# AgriShield Real-Data Preparation

Use this notebook before model training when you have better rainfall, storm-track, or observed-impact data.

Expected required files in a Kaggle dataset:

- `advisory_zones.csv`
- `hazard_observations.csv`

Optional files:

- `rainfall_daily.csv`
- `event_labels.csv`
- `storm_track_points.csv`

The output folder `/kaggle/working/agrishield_real_data_ready/` can be used as input for the baseline training notebook.

```python
from pathlib import Path
import runpy

SCRIPT_CANDIDATES = [
    Path('/kaggle/input/agrishield-demo/kaggle/agrishield_real_data_prep.py'),
    Path('/kaggle/input/agrishield-real-data-prep/agrishield_real_data_prep.py'),
    Path('/kaggle/working/agrishield_real_data_prep.py'),
    Path('agrishield_real_data_prep.py'),
    Path('kaggle/agrishield_real_data_prep.py'),
]

script_path = next((path for path in SCRIPT_CANDIDATES if path.exists()), None)
if script_path is None:
    raise FileNotFoundError(
        'Upload agrishield_real_data_prep.py beside this notebook or attach the repo folder as a Kaggle dataset.'
    )

print('Running:', script_path)
runpy.run_path(str(script_path), run_name='__main__')
```

```python
from pathlib import Path
import json
import pandas as pd

ready_dir = Path('/kaggle/working/agrishield_real_data_ready')
report_path = ready_dir / 'data_quality_report.json'

if report_path.exists():
    report = json.loads(report_path.read_text())
    print(json.dumps(report, indent=2))
    display(pd.read_csv(ready_dir / 'hazard_observations.csv').head())
else:
    print('No report found yet. Run the previous cell first.')
```

## Optional: Train Immediately In The Same Kaggle Session

If `agrishield_kaggle_core.py` is also uploaded or attached, run the next cell. It points the baseline model to the enriched folder.

```python
from pathlib import Path
import os
import runpy

CORE_CANDIDATES = [
    Path('/kaggle/input/agrishield-demo/kaggle/agrishield_kaggle_core.py'),
    Path('/kaggle/input/agrishield-core-script/agrishield_kaggle_core.py'),
    Path('/kaggle/working/agrishield_kaggle_core.py'),
    Path('agrishield_kaggle_core.py'),
    Path('kaggle/agrishield_kaggle_core.py'),
]

core_path = next((path for path in CORE_CANDIDATES if path.exists()), None)
if core_path is None:
    print('Skipping training: agrishield_kaggle_core.py was not found.')
else:
    os.environ['AGRISHIELD_INPUT_DIR'] = '/kaggle/working/agrishield_real_data_ready'
    print('Training with:', core_path)
    runpy.run_path(str(core_path), run_name='__main__')
```
